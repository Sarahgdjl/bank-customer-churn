```python
import pandas as pd
from sqlalchemy import create_engine

# Load CSV
df = pd.read_csv('data/Customer-Churn-Records.csv')

# Rename columns to match staging table
df.columns = [
    'row_number', 'customer_id', 'surname', 'credit_score',
    'geography', 'gender', 'age', 'tenure', 'balance',
    'num_of_products', 'has_cr_card', 'is_active_member',
    'estimated_salary', 'exited', 'complain',
    'satisfaction_score', 'card_type', 'point_earned'
]

# SQLAlchemy engine (note the backtick-quoted db name because of the dash)
engine = create_engine('mysql+mysqlconnector://root:@localhost/churn-warehouse')

# Load into staging
df.to_sql('staging_raw', con=engine, if_exists='append', index=False)

print(f"Loaded {len(df)} rows")
engine.dispose()
```
